```python
df= pd.DataFrame([100,90,90,87,76,87,50,50,55,67,74,43,23,30], columns=['marks']).sort_values(by='marks')
df
```

```python
df['average_rank'] = df['marks'].rank(method='average')
df['min_rank'] = df['marks'].rank(method='min')
df['max_rank'] = df['marks'].rank(method='max')
df['first_rank'] = df['marks'].rank(method='first')
df['dense_rank'] = df['marks'].rank(method='dense')
df
```

```python
# By default, the method is â€˜averageâ€™ in the .rank(), and it assigns the average rank to the equal values

#This method does not skip a rank, like the â€˜minâ€™ and â€˜maxâ€™ method

#If the score is same for two or more observations, then the â€˜maxâ€™ method assigns the maximum rank to all the equal scores

#If the score is same for two or more observations, then the â€˜minâ€™ method assigns the minimum rank to all the equal scores
```

# joining DataFrame

```python
data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}
df1 = pd.DataFrame(data1, columns = ['id', 'Feature1', 'Feature2'])

df1
```

```python
data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}
df2 = pd.DataFrame(data2, columns = ['id', 'Feature1', 'Feature2'])

df2

```

```python
df1.set_index('id').join(df2.set_index('id'),on='id',how='inner',lsuffix='_d1',rsuffix='_d2')
```

```python
df1.set_index('id').join(df2.set_index('id'),on='id',how='outer',lsuffix='_d1',rsuffix='_d2')
```

```python
df1.set_index('id').join(df2.set_index('id'),on='id',how='left',lsuffix='_d1',rsuffix='_d2')
```

```python
df1.set_index('id').join(df2.set_index('id'),on='id',how='right',lsuffix='_d1',rsuffix='_d2')
```

```python
###using merge method
```

```python
df_merged = pd.merge(df1, df2, on='id')

df_merged
```

```python
df_merged = pd.merge(df1, df2, on='id',how='outer')

df_merged
```

# crosstab

```python
#This method is used to compute a simple cross-tabulation of two (or more) factors.
#By default, computes a frequency table of the
#factors unless an array of values and an aggregation function are passed.
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
pd.crosstab(df['sex'],df['embark_town'],values=df['fare'],aggfunc='mean')
```
